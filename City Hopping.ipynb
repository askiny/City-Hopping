Then I wrote a function to clean the data.

```python
def data_cleaner(dataframe, new_column_names):

    first_row = dataframe.columns.tolist()
    new_data = [first_row] + dataframe.values.tolist()
    dataframe = pd.DataFrame(new_data, columns=first_row)
    dataframe.rename(columns=new_column_names, inplace=True)

    return dataframe
```

```python
new_airport = {'1':'Airport ID', 'Goroka Airport' : 'Airport Name', 'Goroka':'City', 'Papua New Guinea':'Country', 'GKA':'IATA', 'AYGA':'ICAO', '-6.081689834590001': 'Latitude', '145.391998291':'Longitude', '5282':'Altitude', '10':'Timezone', 'U':'DST', 'Pacific/Port_Moresby':'TZ Timezone', 'airport':'Type', 'OurAirports':'Source'}

airport_data_new = data_cleaner(airport_data, new_airport)

airport_data_new
```

```python
new_route = {'2B':'Airline', '410' : 'Airline ID', 'AER':'Source Airport', '2965':'Source Airport ID', 'KZN':'Destination Airport', '2990':'Destination Airport ID', 'Unnamed: 6': 'Codeshare', '0':'Stops', 'CR2':'Equipments'}

route_data_new = data_cleaner(route_data, new_route)

route_data_new
```

Then I merged the airport and route data twice to have both destination and source cities-airports match.

```python
airport_data = airport_data_new.drop(columns = ['ICAO', 'Latitude', 'Longitude', 'Altitude', 'Timezone', 'DST', 'TZ Timezone', 'Type', 'Source'])
route_data = route_data_new.drop(columns = ['Codeshare', 'Equipments'])

merged_data = pd.merge(route_data, airport_data, left_on = "Source Airport", right_on = "IATA", how = "left")
merged_data
```

```python
merged_data = pd.merge(merged_data, airport_data, left_on = "Destination Airport", right_on = "IATA", how = "left")
merged_data
```

```python
df_merged_data = pd.DataFrame(merged_data[["Airport Name_x", "City_x", "Airport Name_y", "City_y", 'Stops']])
df_merged_data.rename(columns = {"Airport Name_x":"Source Airport", "City_x":"Source City", "Airport Name_y":"Destination Airport", "City_y":"Destination City"}, inplace = True)
df_merged_data
```

And here is the edges we want:

```python
edges = []

for index, row in df_merged_data.iterrows():
    source_city = row["Source City"]
    destination_city = row["Destination City"]

    if pd.notna(source_city) and pd.notna(destination_city):
        edges.append((source_city, destination_city))
edges[:15]
```

We need shortest paths for Antalya-Deer Lake, Adana-Auckland, Ankara-Kona and Sydney-Churchill so I used shortest_path function from networkx library.

```python
G = nx.from_edgelist(set(edges))

routes = {
    ('Antalya', 'Deer Lake'): nx.shortest_path(G, source='Antalya', target='Deer Lake'),
    ('Adana', 'Auckland'): nx.shortest_path(G, source='Adana', target='Auckland'),
    ('Ankara', 'Kona'): nx.shortest_path(G, source='Ankara', target='Kona'),
    ('Sydney', 'Churchhill'): nx.shortest_path(G, source='Sydney', target='Churchill')
}

for route, path in routes.items():
    print(f"From {route[0]} to {route[1]} route: {path}")
```

And for Page Ranks, I used pagerank funcitons.

```python
pagerank_scores = nx.pagerank(G)

top_cities = sorted(pagerank_scores.items(), key=lambda item: -item[1])[:10]

for city, score in top_cities:
    print(f"{city} : {score}")
```
